```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Leer csv 

df = pd.read_csv('C:/Users/bebit/Desktop/tareas/csv/HousingData.csv')
```

```python
#Muestro columnas

print(df.columns)
```

```python
# Renombro campos
df.rename(columns={'RM':'cantidad_media_habitaciones', 
                   'CRIM':'indice_crimen', 
                   'LSTAT':'porcentaje_clase_baja', 
                   'MEDV':'valor_medio_casas', 
                   'TOWN':'ciudad'}, inplace=True)
```

```python
# Histograma de cantidad media de habitaciones

plt.hist(df['cantidad_media_habitaciones'], bins=20)
plt.title('Distribución de cantidad media de habitaciones')
plt.xlabel('Cantidad media de habitaciones')
plt.ylabel('Frecuencia')
plt.show()

```

```python
# Gráfico de dispersión de índice de crimen vs. valor medio de casas

plt.scatter(df['indice_crimen'], df['valor_medio_casas'])
plt.title('Relacion entre indice de crimen y valor medio de casas')
plt.xlabel('Indice de crimen')
plt.ylabel('Valor medio de casas')
plt.show()

```

```python

# Diagrama de cajas de índice de crimen en diferentes cuantiles de valor medio de casas

df['cuantil'] = pd.qcut(df['valor_medio_casas'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
df.boxplot(column='indice_crimen', by='cuantil')
plt.title('Indice de crimen en diferentes cuantiles de valor medio de casas')
plt.suptitle('')
plt.xlabel('Cuantiles de valor medio de casas')
plt.ylabel('Indice de crimen')
plt.show()
```
