```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np


# DATOS CLINICA

metros = [1, 2, 3, 4, 5, 6, 7, 8]
ninos = [2, 6, 10, 5, 10, 3, 2, 2]
datos = pd.DataFrame({'Metros': metros, 'Niños': ninos})
```

```python
#Tabla de Frecuencia. Diagrama de Barras para las frecuencias absolutas, relativas y acumuladas.

# a) FRECUENCIA
frec_abs = pd.Series(datos['Niños'], index=datos['Metros'])
frec_rel = frec_abs / frec_abs.sum()
frec_acum = frec_rel.cumsum()
tabla_freq = pd.DataFrame({'Frecuencia Absoluta': frec_abs,
                           'Frecuencia Relativa': frec_rel,
                           'Frecuencia Acumulada': frec_acum})
print('Tabla de Frecuencia')
print(tabla_freq)
```

```python
# DIAGRAMA DE BARRAS, RELATIVA Y ACUMULADA

fig, axs = plt.subplots(3, 1, figsize=(8, 10))
tabla_freq['Frecuencia Absoluta'].plot(kind='bar', ax=axs[0])
axs[0].set_title('Frecuencia Absoluta')
tabla_freq['Frecuencia Relativa'].plot(kind='bar', ax=axs[1])
axs[1].set_title('Frecuencia Relativa')
tabla_freq['Frecuencia Acumulada'].plot(kind='bar', ax=axs[2])
axs[2].set_title('Frecuencia Acumulada')
plt.tight_layout()
plt.show()
```

```python
# Mediana, media aritmética, moda y cuartiles

mediana = np.median(datos['Metros'])
media = np.mean(datos['Metros'])
moda = datos['Metros'].mode()[0]
q1 = np.percentile(datos['Metros'], 25)
q3 = np.percentile(datos['Metros'], 75)
print('Mediana:', mediana)
print('Media Aritmética:', media)
print('Moda:', moda)
print('Primer Cuartil (Q1):', q1)
print('Tercer Cuartil (Q3):', q3)
```

```python
# Varianza y desviación estándar
varianza = np.var(datos['Metros'], ddof=1)
desv_estandar = np.std(datos['Metros'], ddof=1)
print('Varianza:', varianza)
print('Desviación Estándar:', desv_estandar)
```

```python
# d) ¿Entre qué dos valores se encuentra, como mínimo, el 75% de las observaciones?

rango_75 = np.percentile(datos['Metros'], [25, 75])
print('Entre estos dos valores se encuentra el 75% de las observaciones:', rango_75)
```
