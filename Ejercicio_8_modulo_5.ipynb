```python
#4. Entregue el gráfico de cada uno de los modelos para los diferentes kernel.


X = iris_df.drop('target', axis=1).values
y = iris.target

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
z_min, z_max = X[:, 2].min() - 1, X[:, 2].max() + 1
w_min, w_max = X[:, 3].min() - 1, X[:, 3].max() + 1

xx, yy, zz, ww = np.meshgrid(
    np.arange(x_min, x_max, 0.25),
    np.arange(y_min, y_max, 0.25),
    np.arange(z_min, z_max, 0.25),
    np.arange(w_min, w_max, 0.25)
)

models = []
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(X, y)
    models.append(model)

# Graficar los modelos para los diferentes kernels
for kernel, model in zip(kernels, models):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Ajustar los límites de los ejes
    ax.set_xlim3d(x_min, x_max)
    ax.set_ylim3d(y_min, y_max)
    ax.set_zlim3d(z_min, z_max)

    # Crear una cuadrícula para evaluar el modelo
    xx = np.linspace(x_min, x_max, 100)
    yy = np.linspace(y_min, y_max, 100)
    ZZ, WW = np.meshgrid(xx, yy)
    xy = np.vstack([ZZ.ravel(), WW.ravel()]).T
    Z = model.predict(xy).reshape(ZZ.shape)

    # Graficar las regiones de decisión y los vectores de soporte
    ax.contour3D(ZZ, WW, Z, alpha=0.5, cmap='viridis')
    ax.scatter3D(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', edgecolors='k')
    ax.scatter3D(
        model.support_vectors_[:, 0], model.support_vectors_[:, 1], model.support_vectors_[:, 2],
        s=100, linewidth=1, facecolors='none', edgecolors='k'
    )
    plt.title(f"SVM con kernel {kernel}")
    plt.xlabel("Sepal Length (cm)")
    plt.ylabel("Sepal Width (cm)")
    plt.show()


```

```python
# 4. Entregue el gráfico de cada uno de los modelos para los diferentes kernel.

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Crear modelos SVM con diferentes kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    # Entrenar el modelo
    svm = SVC(kernel=kernel, C=1.0, random_state=42)
    svm.fit(X_train, y_train)

    # Predecir etiquetas
    y_pred = svm.predict(X_test)

    # Calcular precisión
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy with {kernel} kernel: {accuracy:.2f}")

    # Calcular matriz de confusión
    cm = confusion_matrix(y_test, y_pred)

    # Crear gráfico de matriz de confusión
    fig, ax = plt.subplots(figsize=(8, 9))
    sns.heatmap(cm, annot=True, fmt='d', ax=ax, cmap=plt.cm.Greens, cbar=False)
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title(f'Confusion Matrix ({kernel} kernel)')
    ax.xaxis.set_ticklabels(iris.target_names)
    ax.yaxis.set_ticklabels(iris.target_names)
    plt.show()
```
