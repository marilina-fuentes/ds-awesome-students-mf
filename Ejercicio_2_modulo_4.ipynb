```python
#Caso 1: variable Aleatoria en Distribución normal

import numpy as np
import matplotlib.pyplot as plt

mu = 0  # media
sigma = 1  # desviación estándar

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = np.exp(-(x - mu)**2 / (2*sigma**2)) / (sigma * np.sqrt(2*np.pi))

plt.plot(x, y)
plt.title('Distribución normal')
plt.xlabel('Valores de x')
plt.ylabel('Densidad de probabilidad')
plt.show()
```

```python
#Caso 2: Distribución normal, media y desviación típica

import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros de la distribución normal
mu = 4  # media
sigma = 2  # desviación típica

# Generamos una muestra de valores aleatorios de la distribución normal
num_samples = 1000
samples = np.random.normal(mu, sigma, num_samples)

# Graficamos un histograma de los valores aleatorios generados
plt.hist(samples, bins=30, density=True, alpha=0.5, color='blue')

# Generamos una serie de valores de x para graficar la función de densidad de probabilidad
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Graficamos la función de densidad de probabilidad
plt.plot(x, y, color='red')

# Ajustamos los ejes y agregamos etiquetas y título
plt.xlim(mu - 4*sigma, mu + 4*sigma)
plt.xlabel('Valores de x')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal con media 4 y desviación típica 2')


a=1.666
plt.axvline(x=4-a, color='green', linestyle='--')
plt.axvline(x=4+a, color='green', linestyle='--')




# Mostramos el gráfico
plt.show()
```

```python
# Caso 3: Distribución Normal
# Dada una variable aleatoria continua Z, con distribución normal estándar, es decir, N(0;1),
# encuentre las siguientes probabilidades, usando la tabla.

import pandas as pd
from scipy.stats import norm
import numpy as np


# Creamos un DataFrame con los valores de Z y las probabilidades acumuladas
z_values = pd.DataFrame({'Z': np.arange(0, 3.01, 0.2)})
z_values['Probabilidad'] = norm.cdf(z_values['Z'])

# a) P(0 ≤ Z ≤ 1.25)
prob_a = norm.cdf(1.25) - norm.cdf(0)
print(f"a) P(0 ≤ Z ≤ 1.25) = {prob_a:.4f}")

# b) P(Z ≥ 1.25)
prob_b = 1 - norm.cdf(1.25)
print(f"b) P(Z ≥ 1.25) = {prob_b:.4f}")

# c) P(Z ≤ -1.25)
prob_c = norm.cdf(-1.25)
print(f"c) P(Z ≤ -1.25) = {prob_c:.4f}")

# d) P(0 ≤ Z ≤ 1.33)
prob_d = norm.cdf(1.33) - norm.cdf(0)
print(f"d) P(0 ≤ Z ≤ 1.33) = {prob_d:.4f}")

# e) P(Z ≥ 1.33)
prob_e = 1 - norm.cdf(1.33)
print(f"e) P(Z ≥ 1.33) = {prob_e:.4f}")

# f) P(-1.33 ≤ Z ≤ 0)
prob_f = norm.cdf(0) - norm.cdf(-1.33)
print(f"f) P(-1.33 ≤ Z ≤ 0) = {prob_f:.4f}")

# Mostramos los valores de Z y las probabilidades acumuladas
print("\nValores de Z y probabilidades acumuladas:")
print(z_values)
```

```python
# Caso 4: Distribución Binomial.
# Uno de los ítems de una prueba sorpresa consta de 12 preguntas de verdadero o falso.
# Suponiendo que los estudiantes no saben contestar a ninguna de ellas y, contestan al
# azar, hallar la probabilidad de obtener seis aciertos.

from scipy.stats import binom

# Definimos los parámetros de la distribución binomial
n = 12  # número de preguntas
p = 0.5  # probabilidad de acertar

# Calculamos la probabilidad de obtener 6 aciertos
prob = binom.pmf(6, n, p)
print(f"La probabilidad de obtener 6 aciertos es {prob:.4f}")
```
