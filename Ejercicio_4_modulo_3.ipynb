```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# leer archivo csv
titanic = pd.read_csv('C:/Users/bebit/Desktop/tareas/csv/titanic.csv')
#mostrarlo
print(titanic)
```

```python
#1. Confeccione un gráfico de barras para visualizar hombres y mujeres que han sobrevivido.

# Cantidad de hombres y mujeres que sobrevivieron
survived_by_gender = titanic.groupby(['Sex', 'Survived'])['PassengerId'].count().reset_index()

# creo un grafico de barras 
sns.barplot(x='Sex', y='PassengerId', hue='Survived', data=survived_by_gender)

# modifico titulo y ejes
plt.title('Sobrevivientes por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de pasajeros')

# Mostrar el gráfico
plt.show()
```

```python
#2- # Crear el gráfico de barras y especificar el color

# Agrupar los datos por género y si sobrevivieron o no
survived_by_gender = titanic.groupby(['Sex', 'Survived'])['PassengerId'].count().reset_index()

# Crear un gráfico de barras agrupado para visualizar los resultados
sns.barplot(x='Sex', y='PassengerId', hue='Survived', data=survived_by_gender, palette='Paired')

# Personalizar el gráfico
plt.title('Sobrevivientes por género en otro color')
plt.xlabel('Género')
plt.ylabel('Cantidad de pasajeros')

# Mostrar el gráfico
plt.show()

```

```python
#3- grafico barras agrupados

# Agrupar los datos por género y si sobrevivieron o no
survived_by_gender = titanic.groupby(['Sex', 'Survived'])['PassengerId'].count().reset_index()

# Crear un gráfico de barras agrupado para visualizar los resultados
sns.barplot(x='Sex', y='PassengerId', hue='Survived', data=survived_by_gender)

# Personalizar el gráfico
plt.title('Sobrevivientes por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de pasajeros')

# Mostrar el gráfico
plt.show()
```

```python
#Librería Seaborn.
#Agrupar barras en parcelas es una operación común. Supongamos que desea comparar algunos datos comunes, como la tasa de supervivencia de los pasajeros, pero le gustaría
#agruparlos con algunos criterios.

# Cargar el conjunto de datos Titanic
titanic = sns.load_dataset('titanic')

# Crear un gráfico de barras agrupado por clase y género
sns.catplot(x='sex', y='survived', hue='class', kind='bar', data=titanic, col='embark_town', palette='Paired')
```

```python
#Podríamos querer visualizar la relación de los pasajeros que sobrevivieron, segregados en clases (primera, segunda y tercera), pero también factorizar en qué ciudad se embarcaron.

# Cargo el conjunto de datos Titanic
titanic = sns.load_dataset('titanic')

# Creo gráfico de barras agrupado por clase y factorizado por ciudad de embarque
sns.catplot(x='class', y='survived', hue='embark_town', kind='bar', data=titanic, palette='Paired', col=None)
```

```python
#Esta es una buena cantidad de información en un diagrama y se puede poner fácilmente en un diagrama de barras simple.

# Cargo datos Titanic
titanic = sns.load_dataset('titanic')

# Agrupo datos por clase y ciudad de embarque y calcular la tasa de supervivencia media
survival_rates = titanic.groupby(['class', 'embark_town'])['survived'].mean().reset_index()

# Creo gráfico de barras simple de la tasa de supervivencia por clase y ciudad de embarque
sns.barplot(x='class', y='survived', hue='embark_town', data=survival_rates, palette='Paired')
```

```python
# Para agrupar las barras, usamos el hue argumento. Técnicamente, como su nombre lo indica, el hue, este argumento le dice a Seaborn cómo colorear las barras, pero en el
#proceso de colorear, agrupa datos relevantes.

sns.barplot(x='sex', y='survived', hue='class', data=titanic, palette='Paired')
```
